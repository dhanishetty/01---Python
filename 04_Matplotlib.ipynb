```python
import pandas as pd
from collections import Counter

data = pd.read_csv('data.csv')

ids= data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

languages =[]
popularity = []

for response in lang_responses:
    language_counter.update(response.split(';'))

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)

plt.barh(languages, popularity, color = "#2DDAF8FF")
plt.title("Most popular Languages")
plt.ylabel("Popular Languages")
plt.xlabel("No of People")
plt.show()


```

```python
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity, color = "#2DDAF8FF")
plt.title("Most popular Languages")
plt.ylabel("Popular Languages")
plt.xlabel("No of People")
plt.show()
```

### 5: Filling Area on Line Plots

```python
from matplotlib import pyplot as plt

data = pd.read_csv('05-data.csv')
ages = data['Age' ]
dev_salaries = data['All_Devs' ]
py_salaries = data['Python' ]
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',linestyle = '--', label='All Devs', linewidth = 1)

plt.plot(ages, py_salaries, label='Python', linewidth = 1)


plt.fill_between(ages, py_salaries, alpha = 0.25)#alpha controls transparency

# overall_median = 57287

plt. legend ()

plt.title('Median Salary (USD) by Age')
plt.xlabel ('Ages')
plt. ylabel ('Median Salary (USD)')
plt.show()
```

```python

plt.plot(ages, dev_salaries, color='#444444',linestyle = '--', label='All Devs', linewidth = 1)

plt.plot(ages, py_salaries, label='Python', linewidth = 1)

overall_median = 57287

plt. fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)

plt. fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, color='red', alpha=0.25)

plt. legend ()

plt.title('Median Salary (USD) by Age')
plt.xlabel ('Ages')
plt. ylabel ('Median Salary (USD)')
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',linestyle = '--', label='All Devs', linewidth = 1)

plt.plot(ages, py_salaries, label='Python', linewidth = 1)

overall_median = 57287

plt. fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, alpha=0.25, label = 'Above Avg')

plt. fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True, color='red', alpha=0.25, label = 'Below Avg')

plt. legend ()

plt.title('Median Salary (USD) by Age')
plt.xlabel ('Ages')
plt. ylabel ('Median Salary (USD)')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random
from itertools import count

# Initialize data and counter
x_vals = []
y_vals = []
index = count()

# Create figure and axes explicitly
fig, ax = plt.subplots()

# Animation function
def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))

    ax.clear()  # Clear only this subplot
    ax.plot(x_vals, y_vals)
    ax.set_title("Live Random Data")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    fig.tight_layout()

# Start animation
ani = FuncAnimation(fig, animate, interval=1000, cache_frame_data=False)

# Show plot
plt.show()

```
