# how to start a New topic

1. Start with Basics
    1. What it is?
    2. How it's represented in code ?
    3. Different Operation you can perform on it
    4. Time / Space complexities
    5. Example - Binary Search tree
        1. Inserting a Node
        2. Deleting a Node
        3. Searching a Node 
2. Learn Real-World Applications of data structure & Algorithm- google or chatGPT
3. Use pen & Paper- To visualize
4. implement it from Scratch
    example - When you are working with Stacks try to implement using arrays and linked list. Implement Push, pop and Peek operations.
5. Learn in-built Libraries - lists, sets & Dictionaries
6. Solve easy Problems
