# 184. Department Highest Salary

```python
import pandas as pd

# Sample data
employee_data = {
    "Id": [1, 2, 3, 4, 5],
    "Name": ["Joe", "Jim", "Henry", "Sam", "Max"],
    "Salary": [70000, 90000, 80000, 60000, 90000],
    "DepartmentId": [1, 1, 2, 2, 1]
}

department_data = {
    "Id": [1, 2],
    "Name": ["IT", "Sales"]
}
```

```python
# Create DataFrames
Employee = pd.DataFrame(employee_data)
Department = pd.DataFrame(department_data)

# Merge the data on DepartmentId
df9 = Employee.merge(Department, left_on='DepartmentId', right_on='Id', suffixes=('_Emp', '_Dept'))
df9

```

```python
# Group by department and get max salary
max_salary = df9.groupby('Name_Dept')['Salary'].transform('max')
max_salary
```

```python
# Filter rows where salary equals the max salary in that department
result = df9[df9['Salary'] == max_salary][['Name_Dept', 'Name_Emp', 'Salary']]
print(result)
```

```python
result.columns = ['Department', 'Employee', 'Salary']

print(result)
```
